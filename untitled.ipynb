```python
# Import necessary packages
import pandas as pd
import numpy as np
```

```python
airbnb_price = pd.read_csv('data/airbnb_price.csv')
airbnb_room_type = pd.read_excel('data/airbnb_room_type.xlsx')
airbnb_last_review = pd.read_csv('data/airbnb_last_review.tsv', delimiter='\t')
print("Airbnb Price \n", airbnb_price.head())
print("Airbnb Room Type \n", airbnb_room_type.head())
print("Airbnb Last Review \n", airbnb_last_review.head())
```

```python
df_1 = airbnb_price.merge(airbnb_room_type, on='listing_id')
df = df_1.merge(airbnb_last_review, on='listing_id')
df
```

```python
df.info()
```

```python
df_copy = df.copy()
```

```python
df_copy['last_review'] = pd.to_datetime(df_copy['last_review'], format='%B %d %Y')
print(f"The earliest and moste recent review dates are {df_copy['last_review'].min()} and {df_copy['last_review'].max()}, respectively.")
first_reviewed = df_copy['last_review'].min()
last_reviewed = df_copy['last_review'].max()
```

```python
df_copy['room_type'] = df_copy['room_type'].str.lower()
print(f"The number of listed private rooms is {df_copy[df_copy['room_type'] == 'private room'].shape[0]}.")
nb_private_rooms = df_copy[df_copy['room_type'] == 'private room'].shape[0]
```

```python
df_copy.info()
```

```python
print(df_copy['price'].dtypes)
df_copy['price'] = df_copy['price'].str.replace(' dollars', '').astype('int32')
print(df_copy['price'].dtypes)
print(f"The average price of listings is {round(df_copy['price'].mean(), 2)}.")
avg_price = df_copy['price'].mean()
```

```python
# Handle missing or irregular values when splitting
df_copy[['borough', 'neighborhood']] = df_copy['nbhood_full'].str.split(', ', expand=True, n=1)

# Check for rows where the split failed
missing_boroughs = df_copy[df_copy['borough'].isna() | df_copy['neighborhood'].isna()]
if not missing_boroughs.empty:
    print("Rows with missing borough or neighborhood after split:")
    print(missing_boroughs)

# Calculate average price per night for each borough and neighborhood
average_prices = df_copy.groupby(['borough', 'neighborhood'])['price'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt
# Aggregate by borough only
average_prices_borough = df_copy.groupby('borough')['price'].mean().reset_index()

# Plot the average nightly price per borough
plt.figure(figsize=(8, 6))
plt.bar(average_prices_borough['borough'], average_prices_borough['price'], color='skyblue')

# Add labels and title
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Average Price (Dollars)', fontsize=12)
plt.title('Average Nightly Price by Borough', fontsize=14)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Calculate average prices by room type using df_copy
average_prices_room_type_copy = df_copy.groupby('room_type')['price'].mean().reset_index()

# Plot the average nightly price by room type
plt.figure(figsize=(8, 6))
plt.bar(average_prices_room_type_copy['room_type'], average_prices_room_type_copy['price'], color='lightgreen')

# Add labels and title
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Average Price (Dollars)', fontsize=12)
plt.title('Average Nightly Price by Room Type', fontsize=14)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Calculate the average price per neighborhood
average_prices_neighborhood = df_copy.groupby('neighborhood')['price'].mean().reset_index()

# Identify neighborhoods with the highest and lowest average prices
top_neighborhoods = average_prices_neighborhood.sort_values(by='price', ascending=False).head(5)
bottom_neighborhoods = average_prices_neighborhood.sort_values(by='price', ascending=True).head(5)

top_neighborhoods
```

```python
bottom_neighborhoods
```

```python
# Calculate the proportion of room types within each borough
room_type_proportions = df_copy.groupby(['borough', 'room_type']).size().reset_index(name='count')
total_per_borough = room_type_proportions.groupby('borough')['count'].sum().reset_index(name='total')
room_type_proportions = room_type_proportions.merge(total_per_borough, on='borough')
room_type_proportions['proportion'] = room_type_proportions['count'] / room_type_proportions['total']

# Creating a grouped bar chart
plt.figure(figsize=(12, 8))

# Iterate over unique room types to plot grouped bars
room_types = room_type_proportions['room_type'].unique()
x_indexes = range(len(room_type_proportions['borough'].unique()))
width = 0.2  # Width of each bar

for i, room_type in enumerate(room_types):
    subset = room_type_proportions[room_type_proportions['room_type'] == room_type]
    plt.bar(
        [x + i * width for x in x_indexes],
        subset['proportion'],
        width=width,
        label=room_type,
        alpha=0.7
    )

# Set labels and title
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Proportion of Room Type', fontsize=12)
plt.title('Proportion of Room Types in Each Borough (Grouped Bar Chart)', fontsize=14)
plt.xticks([x + (len(room_types) - 1) * width / 2 for x in x_indexes], room_type_proportions['borough'].unique())
plt.legend(title='Room Type')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Calculate the number of listings per neighborhood
listings_per_neighborhood = df_copy.groupby('neighborhood').size().reset_index(name='count')

# Merge with average price data
listings_and_prices = listings_per_neighborhood.merge(
    df_copy.groupby('neighborhood')['price'].mean().reset_index(name='average_price'),
    on='neighborhood'
)

# Identify neighborhoods with the highest and lowest number of listings
top_listings = listings_and_prices.sort_values(by='count', ascending=False).head(5)
bottom_listings = listings_and_prices.sort_values(by='count', ascending=True).head(5)

top_listings
```

```python
bottom_listings
```

```python
# Plot the top neighborhoods with the highest number of listings (only count)
plt.figure(figsize=(12, 6))
plt.bar(top_listings['neighborhood'], top_listings['count'], color='blue', label='Number of Listings')
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.title('Top 5 Neighborhoods with the Highest Number of Listings', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Count the number of listings in each borough
listings_per_borough = df_copy.groupby('borough').size().reset_index(name='count')

# Plot the distribution of listings across boroughs
plt.figure(figsize=(10, 6))
plt.bar(listings_per_borough['borough'], listings_per_borough['count'], color='purple', alpha=0.7)
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.title('Distribution of Listings Across Boroughs', fontsize=14)
plt.tight_layout()
plt.show()

```

```python
# Extract the month from the 'last_review' column
df_copy['review_month'] = df_copy['last_review'].dt.month

# Count the number of reviews per month for each borough
seasonal_trends_borough = df_copy.groupby(['borough', 'review_month']).size().reset_index(name='review_count')

# Plot seasonal trends in the number of reviews for each borough
plt.figure(figsize=(12, 8))

for borough in seasonal_trends_borough['borough'].unique():
    subset = seasonal_trends_borough[seasonal_trends_borough['borough'] == borough]
    plt.plot(subset['review_month'], subset['review_count'], marker='o', label=borough)

plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.title('Seasonal Trends in Number of Last Reviews by Borough', fontsize=14)
plt.xticks(range(1, 13), [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.legend(title='Borough')
plt.tight_layout()
plt.show()

```

```python
# Count the number of reviews per month for each neighborhood
seasonal_trends_neighborhood = df_copy.groupby(['neighborhood', 'review_month']).size().reset_index(name='review_count')

# Plot seasonal trends in the number of reviews for the top 5 neighborhoods
top_neighborhoods = seasonal_trends_neighborhood.groupby('neighborhood')['review_count'].sum().nlargest(5).index
top_neighborhood_trends = seasonal_trends_neighborhood[seasonal_trends_neighborhood['neighborhood'].isin(top_neighborhoods)]

plt.figure(figsize=(12, 8))

for neighborhood in top_neighborhood_trends['neighborhood'].unique():
    subset = top_neighborhood_trends[top_neighborhood_trends['neighborhood'] == neighborhood]
    plt.plot(subset['review_month'], subset['review_count'], marker='o', label=neighborhood)

plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.title('Seasonal Trends in Number of Last Reviews for Top Neighborhoods', fontsize=14)
plt.xticks(range(1, 13), [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.legend(title='Neighborhood')
plt.tight_layout()
plt.show()
```

```python
# Extract year and month from the 'last_review' column for trend analysis
df_copy['review_year_month'] = df_copy['last_review'].dt.to_period('M')

# Count the number of reviews per room type over time
reviews_trend_room_type = df_copy.groupby(['room_type', 'review_year_month']).size().reset_index(name='review_count')

# Plot the number of reviews over time for each room type
plt.figure(figsize=(14, 8))
for room_type in reviews_trend_room_type['room_type'].unique():
    subset = reviews_trend_room_type[reviews_trend_room_type['room_type'] == room_type]
    plt.plot(subset['review_year_month'].astype(str), subset['review_count'], marker='o', label=room_type)

plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.title('Number of Reviews Over Time for Different Room Types', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()


```

```python
# Count the number of reviews per neighborhood over time
reviews_trend_neighborhood = df_copy.groupby(['neighborhood', 'review_year_month']).size().reset_index(name='review_count')

# Focus on the top 5 neighborhoods with the most reviews for clarity
top_reviewed_neighborhoods = reviews_trend_neighborhood.groupby('neighborhood')['review_count'].sum().nlargest(5).index
top_neighborhood_trends = reviews_trend_neighborhood[reviews_trend_neighborhood['neighborhood'].isin(top_reviewed_neighborhoods)]

# Plot the number of reviews over time for the top neighborhoods
plt.figure(figsize=(14, 8))
for neighborhood in top_neighborhood_trends['neighborhood'].unique():
    subset = top_neighborhood_trends[top_neighborhood_trends['neighborhood'] == neighborhood]
    plt.plot(subset['review_year_month'].astype(str), subset['review_count'], marker='o', label=neighborhood)

plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.title('Number of Reviews Over Time for Top Neighborhoods', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Neighborhood')
plt.tight_layout()
plt.show()
```

```python
import squarify

# Prepare data for the treemap: total listings and average price by neighborhood
treemap_data = df_copy.groupby('neighborhood').agg(
    total_listings=('listing_id', 'count'),
    avg_price=('price', 'mean')
).reset_index()

# Sort the data by total listings to emphasize larger neighborhoods
treemap_data = treemap_data.sort_values(by='total_listings', ascending=False)

# Create labels for the treemap
treemap_data['label'] = treemap_data['neighborhood'] + "\n" + \
                        "Listings: " + treemap_data['total_listings'].astype(str) + "\n" + \
                        "Avg Price: $" + treemap_data['avg_price'].round(2).astype(str)


```

```python
# Select the top 20 neighborhoods by total listings
top_20_treemap_data = treemap_data.head(20)

# Create labels for the treemap
top_20_treemap_data['label'] = top_20_treemap_data['neighborhood'] + "\n" + \
                               "Listings: " + top_20_treemap_data['total_listings'].astype(str) + "\n" + \
                               "Avg Price: $" + top_20_treemap_data['avg_price'].round(2).astype(str)

# Plot the treemap for the top 20 neighborhoods
plt.figure(figsize=(14, 8))
squarify.plot(
    sizes=top_20_treemap_data['total_listings'],
    label=top_20_treemap_data['label'],
    color=plt.cm.viridis(top_20_treemap_data['avg_price'] / top_20_treemap_data['avg_price'].max()),
    alpha=0.8
)
plt.title('Treemap of Listings and Prices for Top 20 Neighborhoods', fontsize=16)
plt.axis('off')
plt.show()
```

```python
# Calculate the count of each room type within each borough
room_type_borough = df_copy.groupby(['borough', 'room_type']).size().reset_index(name='count')

# Pivot the data for a grouped bar chart
room_type_borough_pivot = room_type_borough.pivot(index='borough', columns='room_type', values='count').fillna(0)

# Plot the grouped bar chart
room_type_borough_pivot.plot(kind='bar', figsize=(12, 8), width=0.8)

# Add labels and title
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.title('Room Types Across Boroughs', fontsize=14)
plt.legend(title='Room Type')
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()

# Display the chart
plt.show()
```
